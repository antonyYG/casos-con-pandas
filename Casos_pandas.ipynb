```python
import pandas as pd
from datetime import datetime
```

```python
caso1 = pd.read_excel('caso_1_ventas_tienda.xlsx')
```

```python
caso1
```

```python
caso1['total'] = caso1['price'] * caso1['quantity']
caso1.groupby('category')['total'].sum()
```

```python
caso1.groupby('product')['total'].sum()
```

```python
caso1['customer_id'].value_counts().head(2)
```

```python
caso1.groupby('date')['quantity'].sum().sort_values(ascending=False).head(1)
```

```python
caso1['mes'] = caso1['date'].dt.to_period('M')
caso1.groupby('mes')['quantity'].sum()
```

```python
caso1.groupby('product')['total'].sum().sort_values(ascending=False).head(1)
```

Caso5

```python
caso5= pd.read_excel('caso_5_reservas_hotel.xlsx')
caso5
```

```python
caso5['dias'] = (caso5['check_out'] - caso5['check_in']).dt.days
caso5
```

```python
caso5['total_price'] = caso5['dias'] * caso5['price_per_night']
caso5
```

```python
caso5.loc[caso5['total_price'].idxmax()]
```

```python
caso5['room_type'].value_counts().head(1)
```

```python
caso5['dias'].mean()
```

```python
caso5['customer_name'].count()
```

```python
caso5.groupby('room_type')['total_price'].sum()
```
